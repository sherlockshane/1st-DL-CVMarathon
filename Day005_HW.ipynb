1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests

r = requests.get(' https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response=r.text

#len(response)

import json

data=json.loads(r.text)
len(data)
print('欄位名稱\n',data[0].keys())
#print(data[1].keys())
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

#觀察上面的欄位名稱，把題目要求的找出來print出來
for d in data:
    print(d['title'])
    print(d['createdAt'])
    print(d['commentCount'])
    print(d['likeCount'])
    
    
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

articles=[]
com=0
for d in data:
    com+=d['commentCount']
    articles.append(d['title'])


#print(articles)
#print(len(articles))
length=len(articles)

likes=0

for d in data:
    likes+=d['likeCount']
    #articles=d['title']
    #print(d['title'])
    #print(len(d['title']))
#print(likes)
avglike=likes/length
avgcom=com/length

#------------非熱門-----------------
rr = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response2=rr.text
data2 = json.loads(response2)

articles2=[]
com2=0
for dd in data2:
    com2+=dd['commentCount']
    articles2.append(dd['title'])

length2=len(articles2)    
likes2=0
for dd in data2:
    likes2+=dd['likeCount']

avglike2=likes2/length2
avgcom2=com2/length2

print('熱門平均留言數\n',avgcom)
print('熱門平均按讚數\n',avglike)

print('非熱門平均留言數\n',avgcom2)
print('非熱門平均按讚數\n',avglike2)

```
